```python
from q_align import QAlignVideoScorer, load_video

# Initialize the video scorer
scorer = QAlignVideoScorer()
```

```python
from google.colab import files
import io

# Upload multiple video files
uploaded = files.upload()

# Process each uploaded video
for video_filename, video_content in uploaded.items():
    # Save the uploaded video to a file
    with open(video_filename, "wb") as f:
        f.write(video_content)

    # Load and score the video
    video = load_video(video_filename)
    video_score = scorer([video])

    print(f"Video Quality Score for {video_filename}: {video_score.tolist()}")
```
