```python
import base64
import requests
from PIL import Image
import io
import moviepy.editor as mp
import json


# OpenAI API Key
api_key = "sk-proj-gk7MnSYvas4GMRXqOSMaT3BlbkFJAydBCcyaSiw7JYPNeDvC"

# Function to encode the image
def encode_image(image):
    img_bytes = io.BytesIO()
    image.save(img_bytes, format='PNG')
    img_bytes = img_bytes.getvalue()
    return base64.b64encode(img_bytes).decode('utf-8')
```

```python
def checkgoodness(gif_path):
# Open the GIF file
  with Image.open(gif_path) as im:
      yes_count = 0
      no_count = 0
      # Get the number of frames in the GIF
      num_frames = im.n_frames

      # Iterate over each frame
      for frame_num in range(num_frames):
          # Get the current frame
          im.seek(frame_num)

          # Encode the frame as a base64 string
          base64_image = encode_image(im)

          headers = {
              "Content-Type": "application/json",
              "Authorization": f"Bearer {api_key}"
          }

          payload = {
              "model": "gpt-4o",
              "messages": [
                  {
                      "role": "user",
                      "content": [
                          {
                              "type": "text",
                              "text": "How real does this image look? Give the answer in Yes/No"
                          },
                          {
                              "type": "image_url",
                              "image_url": {
                                  "url": f"data:image/jpeg;base64,{base64_image}"
                              }
                          }
                      ]
                  }
              ],
              "max_tokens": 300
          }

          response = requests.post("https://api.openai.com/v1/chat/completions", headers=headers, json=payload)
          print(f"Frame {frame_num}: {response.json()}")
          result = response.json()["choices"][0]["message"]["content"]

          if "Yes" in result:
              yes_count += 1
          elif "No" in result:
              no_count += 1

      # Determine if the image is good or not based on the majority of responses
      if yes_count > no_count:
          print("The image is good.")
      else:
          print("The image is not good.")
```
