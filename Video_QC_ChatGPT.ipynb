```python
import base64
import requests
import cv2

# OpenAI API Key
api_key = "sk-proj-gk7MnSYvas4GMRXqOSMaT3BlbkFJAydBCcyaSiw7JYPNeDvC"

# Function to encode the image
def encode_image(image):
    _, encoded_image = cv2.imencode(".png", image)
    return base64.b64encode(encoded_image).decode('utf-8')
```

```python
def detectcorrectnessinvideo(video_path):

    # Open the video file
    video_capture = cv2.VideoCapture(video_path)
    yes_count = 0
    no_count = 0
    # Loop through each frame
    while True:
        # Read a frame from the video
        ret, frame = video_capture.read()

        # If there are no more frames, break the loop
        if not ret:
            break

        # Encode the frame as a base64 string
        base64_image = encode_image(frame)

        # Update the payload with the new base64 string
        headers = {"Content-Type": "application/json", "Authorization": f"Bearer {api_key}"}
        payload = {
            "model": "gpt-4o",
            "messages": [
                {
                    "role": "user",
                    "content": [

                        {"type": "text", "text": "How real does this image look? Give the answer in Yes/No"},
                        {"type": "image_url", "image_url": {"url": f"data:image/jpeg;base64,{base64_image}"}},
                    ],
                }
            ],
            "max_tokens": 300,
        }

        # Send the request to the OpenAI API
        response = requests.post("https://api.openai.com/v1/chat/completions", headers=headers, json=payload)

        # Process the API response
        print(response.json())
        result = response.json()["choices"][0]["message"]["content"]

        if "Yes" in result:
                yes_count += 1
        elif "No" in result:
                no_count += 1

        # Determine if the image is good or not based on the majority of responses
    if yes_count > no_count:
        print("The image is good.")
    else:
        print("The image is not good.")

    # Release the video capture object
    video_capture.release()
```
